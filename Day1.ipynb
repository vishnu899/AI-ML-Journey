#Python Mini Calculator

```python
num_sub = int(input("How many subjects do you have? "))

marks = []

for i in range(num_sub):
  while True:
    mark_input = int (input(f"Enter marks for subject {(i+1)}: "))
    marks.append(mark_input)
    if mark_input >= 0 and mark_input <= 100:
      break
    else:
      print("Invalid input")
```

```python
print(marks)
```

```python
import builtins

sum_marks = builtins.sum(marks)
total_max_marks = num_sub*100
percentage = (sum_marks/total_max_marks)*100

print(percentage)
```

```python
if percentage >= 90:
  print("Grade: A")
elif percentage >= 80:
  print("Grade: B")
elif percentage >= 70:
  print("Grade: C")
elif percentage >= 60:
  print("Grade: D")
```
