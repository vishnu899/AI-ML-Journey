```python
height = float(input(f"Enter you height in cm: "))
weight = float(input(f"Enter your weight in kg: "))
print(f"Height is {height} and your weight is {weight}")
```

```python
# convert the height to metres
height = height/100
print(height)
```

```python
bmi = weight/(height**2)
print(bmi)
```

```python
if(bmi<18.5):
  print("Underweight")
elif bmi>18.5 and bmi<24.9:
  print("Healthy")
elif bmi>25 and bmi<30:
  print("Overweight")
else:
  print("Obese")
```
