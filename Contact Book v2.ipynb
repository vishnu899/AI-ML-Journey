Build a contact management system with features to add, list, search, update, and delete contacts, along with saving and loading contact data to/from a JSON file, and a command-line interface.

## Define Contact Structure and Initialize Storage

### Subtask:
Establish the data structure for a contact (e.g., name, phone, email) and initialize an in-memory list or dictionary to store contacts. This will also include the initial setup for saving to and loading from a JSON file.


**Reasoning**:
To begin, I will define an empty list for contacts and specify the filename for JSON storage, as instructed.



```python
contacts = []
CONTACTS_FILE = 'contacts.json'

print(f"'contacts' list initialized as empty: {contacts}")
print(f"'CONTACTS_FILE' set to: '{CONTACTS_FILE}'")
```

## Implement Add Contact Feature

### Subtask:
Develop a function to allow users to add new contacts, including inputting contact details and storing them in the defined data structure.


**Reasoning**:
I will define the `add_contact` function as per the instructions, taking user input for name, phone, and email, validating the name, creating a contact dictionary, appending it to the global `contacts` list, and providing a confirmation.



```python
def add_contact():
    while True:
        name = input("Enter contact name (required): ").strip()
        if name:
            break
        else:
            print("Name cannot be empty. Please enter a name.")

    phone = input("Enter phone number (optional): ").strip()
    email = input("Enter email address (optional): ").strip()

    contact = {
        'name': name,
        'phone': phone,
        'email': email
    }
    contacts.append(contact)
    print(f"Contact '{name}' added successfully!")
    print(f"Current contacts: {contacts}")

add_contact()
add_contact()
```

## Implement List Contacts Feature

### Subtask:
Create a function to display all stored contacts. This will initially be a basic listing, which will be enhanced later with 'Pretty print'.


**Reasoning**:
I will define the `list_contacts` function as instructed, which will check if the global `contacts` list is empty, and if not, iterate through it to print each contact's details in a readable format, then call the function to demonstrate its functionality.



```python
def list_contacts():
    if not contacts:
        print("No contacts available.")
        return

    print("\n--- Your Contacts ---")
    for i, contact in enumerate(contacts):
        print(f"Contact {i+1}:")
        print(f"  Name: {contact['name']}")
        print(f"  Phone: {contact['phone']}")
        print(f"  Email: {contact['email']}")
    print("---------------------")

# Demonstrate the function
list_contacts()
```

## Implement Search Contact Feature

### Subtask:
Develop a function to search for contacts based on criteria like name or phone number.


**Reasoning**:
I will define the `search_contact` function as per the instructions, iterating through the global `contacts` list, performing a case-insensitive search on 'name', 'phone', and 'email' fields, and returning a list of matching contacts. I will then demonstrate its usage with sample search terms.



```python
def search_contact(search_term):
    found_contacts = []
    search_term_lower = search_term.lower()

    for contact in contacts:
        if (search_term_lower in contact['name'].lower() or
            search_term_lower in contact['phone'].lower() or
            search_term_lower in contact['email'].lower()):
            found_contacts.append(contact)
    return found_contacts

# Demonstrate the function
print("\n--- Searching for 'Vishnu' ---")
results1 = search_contact('Vishnu')
if results1:
    for i, contact in enumerate(results1):
        print(f"Contact {i+1}: {contact['name']} - {contact['phone']} - {contact['email']}")
else:
    print("No contacts found.")

print("\n--- Searching for '8542' (phone number) ---")
results2 = search_contact('8542')
if results2:
    for i, contact in enumerate(results2):
        print(f"Contact {i+1}: {contact['name']} - {contact['phone']} - {contact['email']}")
else:
    print("No contacts found.")

print("\n--- Searching for 'nonexistent' ---")
results3 = search_contact('nonexistent')
if results3:
    for i, contact in enumerate(results3):
        print(f"Contact {i+1}: {contact['name']} - {contact['phone']} - {contact['email']}")
else:
    print("No contacts found.")
```

## Implement Save and Load to JSON Feature

### Subtask:
Develop functions to save the current list of contacts to a JSON file and load contacts from a JSON file upon application startup. This will make the contact book persistent.


**Reasoning**:
I will import the `json` module, then define the `save_contacts` and `load_contacts` functions with error handling, and finally demonstrate their functionality by saving the current contacts and then loading them back, verifying with `list_contacts`.



```python
import json

def save_contacts(contacts_list, file_path):
    try:
        with open(file_path, 'w') as f:
            json.dump(contacts_list, f, indent=4)
        print(f"Contacts successfully saved to '{file_path}'.")
    except IOError as e:
        print(f"Error saving contacts to '{file_path}': {e}")

def load_contacts(file_path):
    global contacts
    try:
        with open(file_path, 'r') as f:
            contacts = json.load(f)
        print(f"Contacts successfully loaded from '{file_path}'.")
        return contacts
    except FileNotFoundError:
        print(f"'{file_path}' not found. Initializing with an empty contact list.")
        contacts = []
        return contacts
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON from '{file_path}': {e}. Initializing with an empty contact list.")
        contacts = []
        return contacts
    except IOError as e:
        print(f"Error loading contacts from '{file_path}': {e}. Initializing with an empty contact list.")
        contacts = []
        return contacts

# 4. Call save_contacts() to save the currently existing contacts
print("\n--- Saving current contacts ---")
save_contacts(contacts, CONTACTS_FILE)

# 5. Demonstrate loading contacts
print("\n--- Demonstrating loading contacts ---")
print("Clearing current contacts list for demonstration...")
contacts = [] # Clear the list to show loading functionality
print(f"Contacts after clearing: {contacts}")

print("Attempting to load contacts...")
loaded_contacts = load_contacts(CONTACTS_FILE)
print(f"Contacts after loading: {loaded_contacts}")

print("Displaying loaded contacts:")
list_contacts()
```

## Implement Update Contact Feature

### Subtask:
Develop a function to allow users to modify existing contact details.


**Reasoning**:
I will define the `update_contact` function as per the instructions, which involves prompting the user to find a contact, using the `search_contact` function, handling single or multiple search results, asking for updated details while allowing current values to be kept, modifying the contact in the `contacts` list, and then providing demonstrations of its functionality, including updating an existing contact, trying to update a non-existent contact, and updating a contact from multiple search results.



```python
def update_contact():
    search_term = input("Enter name, phone, or email of the contact to update: ").strip()
    if not search_term:
        print("Search term cannot be empty. Returning to menu.")
        return

    found_contacts = search_contact(search_term)

    if not found_contacts:
        print(f"No contacts found matching '{search_term}'.")
        return

    contact_to_update = None
    if len(found_contacts) == 1:
        contact_to_update = found_contacts[0]
        print("Found one contact:")
        display_contact_details(contact_to_update)
    else:
        print(f"Multiple contacts found matching '{search_term}':")
        for i, contact in enumerate(found_contacts):
            print(f"  [{i+1}] Name: {contact['name'].title()}, Phone: {contact['phone']}, Email: {contact['email']}")
        while True:
            try:
                selection = int(input("Enter the number of the contact to update: "))
                if 1 <= selection <= len(found_contacts):
                    contact_to_update = found_contacts[selection - 1]
                    break
                else:
                    print("Invalid selection. Please enter a number from the list.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    if contact_to_update:
        print("\nEnter new details (press Enter to keep current value):")
        current_name = contact_to_update['name']
        current_phone = contact_to_update['phone']
        current_email = contact_to_update['email']

        new_name = input(f"Name (current: {current_name}): ").strip()
        if new_name:
            contact_to_update['name'] = new_name

        new_phone = input(f"Phone (current: {current_phone}): ").strip()
        if new_phone:
            contact_to_update['phone'] = new_phone

        new_email = input(f"Email (current: {current_email}): ").strip()
        if new_email:
            contact_to_update['email'] = new_email

        # Find the index of the contact_to_update and update it in the global contacts list
        # This is important because contact_to_update is a reference to the dict in found_contacts,
        # and found_contacts is a copy, so we need to update the original list.
        for i, contact in enumerate(contacts):
            if contact['name'] == current_name and contact['phone'] == current_phone and contact['email'] == current_email:
                contacts[i] = contact_to_update
                break

        print(f"Contact '{contact_to_update['name']}' updated successfully!")
        print("Updated contact details:")
        display_contact_details(contact_to_update)

# --- Demonstrations of update_contact function ---

print("\n--- Demonstration 1: Update an existing contact (Vishnu's email) ---")
update_contact() # Search for 'Vishnu', update email

print("\n--- Demonstration 2: Try to update a non-existent contact ---")
update_contact() # Search for 'NonExistentPerson'

print("\n--- Add a duplicate contact for multiple search results demo ---")
# Temporarily add a contact to test multiple selection if current contacts don't provide this scenario.
# If 'Vishnu' is already in the list, searching for 'Vishnu' would yield multiple results.
contacts.append({'name': 'Vishnu', 'phone': '1112223333', 'email': 'vishnu.alt@example.com'})
print("Added a duplicate 'Vishnu' for demonstration.")
list_contacts()

print("\n--- Demonstration 3: Update a contact from multiple search results ---")
update_contact() # Search for 'Vishnu', select one to update

print("\n--- Final list of contacts after demonstrations ---")
list_contacts()
```

**Reasoning**:
I will define the `delete_contact` function as per the instructions, which will prompt the user for a search term, use `search_contact` to find matches, handle scenarios with no, single, or multiple results, ask for confirmation before deleting, and then remove the selected contact from the `contacts` list. I will also include demonstrations to show its functionality.



```python
def delete_contact():
    search_term = input("Enter name, phone, or email of the contact to delete: ").strip()
    if not search_term:
        print("Search term cannot be empty. Returning to menu.")
        return

    found_contacts = search_contact(search_term)

    if not found_contacts:
        print(f"No contacts found matching '{search_term}'.")
        return

    contact_to_delete = None
    if len(found_contacts) == 1:
        contact_to_delete = found_contacts[0]
        print("Found one contact:")
        display_contact_details(contact_to_delete)
    else:
        print(f"Multiple contacts found matching '{search_term}':")
        for i, contact in enumerate(found_contacts):
            print(f"  [{i+1}] Name: {contact['name'].title()}, Phone: {contact['phone']}, Email: {contact['email']}")
        while True:
            try:
                selection = int(input("Enter the number of the contact to delete: "))
                if 1 <= selection <= len(found_contacts):
                    contact_to_delete = found_contacts[selection - 1]
                    break
                else:
                    print("Invalid selection. Please enter a number from the list.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    if contact_to_delete:
        confirm = input(f"Are you sure you want to delete '{contact_to_delete['name']}'? (yes/no): ").lower().strip()
        if confirm == 'yes':
            # Find the actual contact in the global contacts list and remove it
            # We need to find its index because contact_to_delete is a copy or reference from search results
            for i, contact in enumerate(contacts):
                if (contact['name'] == contact_to_delete['name'] and
                    contact['phone'] == contact_to_delete['phone'] and
                    contact['email'] == contact_to_delete['email']):
                    del contacts[i]
                    print(f"Contact '{contact_to_delete['name']}' deleted successfully!")
                    break
        else:
            print("Deletion cancelled.")

# --- Demonstrations of delete_contact function ---

print("\n--- Current contacts before delete demos ---")
list_contacts()

print("\n--- Demonstration 1: Delete an existing contact (sagul) ---")
delete_contact() # Search for 'sagul'

print("\n--- Current contacts after first delete demo ---")
list_contacts()

print("\n--- Demonstration 2: Try to delete a non-existent contact ---")
delete_contact() # Search for 'NonExistentPerson'

print("\n--- Demonstration 3: Delete a contact from multiple search results (Vishnu) ---")
delete_contact() # Search for 'Vishnu'

print("\n--- Final list of contacts after all delete demos ---")
list_contacts()
```
