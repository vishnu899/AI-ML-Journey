```python
def my_function(parameter1, parameter2):
    """Docstring: Explain what the function does.

    Args:
        parameter1: Description of parameter1.
        parameter2: Description of parameter2.

    Returns:
        Description of what the function returns.
    """
    # Function body: Your code goes here
    result = parameter1 + parameter2
    return result

# Example usage:
output = my_function(5, 3)
print(output)
```

Let's break down this function template:

*   **`def`**: This keyword indicates the start of a function definition.
*   **`my_function`**: This is the name of the function. Choose a name that describes what the function does.
*   **`(parameter1, parameter2)`**: These are the parameters (inputs) the function accepts. You can have zero or more parameters.
*   **`:`**: A colon marks the end of the function header.
*   **Indentation**: The code inside the function must be indented (usually 4 spaces).
*   **Docstring**: The triple-quoted string right after the `def` line is a docstring. It's a best practice to explain what the function does, its arguments (`Args`), and what it returns (`Returns`).
*   **`# Function body`**: This is where you write the logic of your function.
*   **`return`**: This keyword is used to send a value back from the function. If `return` is not explicitly used, the function implicitly returns `None`.

```python
def add_task(new_task):
  global tasks # Declare that we intend to modify the global 'tasks' list
  tasks.append(new_task)
  return tasks # Optionally return the updated list
```

```python
# Let's add a new task using our corrected function
add_task("Go for a run")
print(tasks)
```

```python
def remove_task(task):
  global tasks
  tasks.remove(task)
  return tasks
```

```python
remove_task('Go for a run')
```

```python
def completed(task):
  global tasks
  tasks.remove(task)
  return tasks
```

```python
completed('Learn Python lists')
```

```python
def display():
  print(tasks)
```

```python
def option_1_action():
    print("You selected Option 1. Performing action...")
    # Add your code for Option 1 here

def option_2_action():
    print("You selected Option 2. Performing another action...")
    # Add your code for Option 2 here

def option_3_action():
    print("You selected Option 3. Doing something else...")
    # Add your code for Option 3 here

def display_menu():
    print("\n--- Menu ---")
    print("1. Option 1")
    print("2. Option 2")
    print("3. Option 3")
    print("4. Exit")
    print("------------")

def main_menu():
    while True:
        display_menu()
        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            option_1_action()
        elif choice == '2':
            option_2_action()
        elif choice == '3':
            option_3_action()
        elif choice == '4':
            print("Exiting the program. Goodbye!")
            break # Exit the while loop
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

# To run the menu:
# main_menu()
```

Here's a breakdown of this menu-driven template:

*   **`option_1_action()`, `option_2_action()`, `option_3_action()`**: These are placeholder functions where you'd put the actual code for each menu option. They simply print a message for now.
*   **`display_menu()`**: This function is responsible for printing the menu options to the console.
*   **`main_menu()`**: This is the core of the menu system:
    *   It uses a `while True:` loop to continuously display the menu and prompt for input until the user chooses to exit.
    *   `input("Enter your choice (1-4): ")` captures the user's choice.
    *   An `if/elif/else` block checks the `choice` and calls the appropriate function. If the choice is '4', it prints an exit message and uses `break` to terminate the loop.
    *   An `else` block handles invalid input.

To run this menu, you would uncomment the last line `# main_menu()` and execute the cell. You can then interact with the menu in the output.

```python
from os import remove
def display_menu():
    print("\n--- Menu ---")
    print("1. Add Tasks")
    print("2. Remove Tasks")
    print("3. Completed Tasks")
    print("4. Display Tasks")
    print("5. Exit")
    print("------------")

def main_menu():
    while True:
        display_menu()
        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            task_input = str(input("Enter a task to add: "))
            add_task(task_input)
        elif choice == '2':
            task_input = str(input("Enter a task to remove: "))
            remove_task(task_input)
        elif choice == '3':
            task_input = str(input("Enter a task to mark as completed: "))
            completed(task_input)
        elif choice == '4':
            display()
        elif choice == '5':
            print("Exiting the program. Goodbye!")
            break # Exit the while loop
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")
```

```python
main_menu()
```
