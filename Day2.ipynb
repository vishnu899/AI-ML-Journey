## Setup Expense Data

### Subtask:
Initialize the `expenses.csv` file by ensuring it exists and has the correct headers: `date,amount,category`. This will prevent errors when adding new entries.


**Reasoning**:
I need to import the `os` and `pandas` libraries to handle file operations and DataFrame creation, which is the first step in initializing the `expenses.csv` file.



```python
import os
import pandas as pd
print("Imported 'os' and 'pandas' libraries.")
```

**Reasoning**:
Now that the libraries are imported, I will define the filename and headers, then check if the 'expenses.csv' file exists. If it doesn't, I will create an empty DataFrame with the specified headers and save it to 'expenses.csv'.

## Add New Expense

### Subtask:
Implement a function that automatically records the current date, prompts the user for the expense amount and category, appends this data to `expenses.csv`, and then calculates and displays the total amount spent so far from all entries in the CSV.


**Reasoning**:
To implement the `add_expense` function, I first need to import the `datetime` module to handle date-related operations, as specified in the instructions.



```python
import datetime
print("Imported 'datetime' module.")
```

**Reasoning**:
Now that the `datetime` module is imported, I will define the `add_expense` function as instructed. This function will get the current date, prompt the user for expense details, create a new DataFrame row, append it to the existing `expenses.csv`, and then calculate and display the total amount spent.



```python
def add_expense():
    today = datetime.date.today().strftime('%Y-%m-%d')

    while True:
        try:
            amount = float(input("Enter amount: "))
            break
        except ValueError:
            print("Invalid amount. Please enter a number.")

    category = input("Enter category: ")

    new_expense_df = pd.DataFrame([{'date': today, 'amount': amount, 'category': category}])

    # Read existing expenses, append new one, and save
    try:
        existing_expenses_df = pd.read_csv('expenses.csv')
    except FileNotFoundError:
        print("Error: 'expenses.csv' not found. Please ensure it's initialized.")
        return

    updated_expenses_df = pd.concat([existing_expenses_df, new_expense_df], ignore_index=True)
    updated_expenses_df.to_csv('expenses.csv', index=False)

    total_spent = updated_expenses_df['amount'].sum()
    print(f"Expense added: Date: {today}, Amount: {amount:.2f}, Category: {category}")
    print(f"Total amount spent: {total_spent:.2f}")

print("Defined 'add_expense' function.")
```

**Reasoning**:
The `add_expense` function has been defined. To fulfill the subtask requirements, I need to call this function to add an actual expense to `expenses.csv` and verify its functionality. This will demonstrate that the function correctly records the date, prompts for input, appends to the CSV, and displays the total amount spent.



```python
add_expense()
print("Called 'add_expense' function to add a new expense.")
```

**Reasoning**:
The `FutureWarning` during `pd.concat` arises from potential inconsistencies in DataFrame dtypes, particularly when `expenses.csv` is initially empty and then read. To address this, I will explicitly define the data types for columns when reading `expenses.csv` within the `add_expense` function, ensuring consistent type handling and preventing the warning.



```python
def add_expense():
    today = datetime.date.today().strftime('%Y-%m-%d')

    while True:
        try:
            amount = float(input("Enter amount: "))
            break
        except ValueError:
            print("Invalid amount. Please enter a number.")

    category = input("Enter category: ")

    new_expense_df = pd.DataFrame([{'date': today, 'amount': amount, 'category': category}])

    # Define dtypes to ensure consistency and avoid FutureWarning
    dtype_mapping = {'date': str, 'amount': float, 'category': str}

    # Read existing expenses, append new one, and save
    try:
        existing_expenses_df = pd.read_csv('expenses.csv', dtype=dtype_mapping)
    except FileNotFoundError:
        print("Error: 'expenses.csv' not found. Please ensure it's initialized.")
        return

    updated_expenses_df = pd.concat([existing_expenses_df, new_expense_df], ignore_index=True)
    updated_expenses_df.to_csv('expenses.csv', index=False)

    total_spent = updated_expenses_df['amount'].sum()
    print(f"Expense added: Date: {today}, Amount: {amount:.2f}, Category: {category}")
    print(f"Total amount spent: {total_spent:.2f}")

print("Redefined 'add_expense' function to prevent FutureWarning.")
```

**Reasoning**:
The `add_expense` function has been redefined to handle the `FutureWarning`. To confirm the fix and ensure the function still works as intended, I will call it again to add another expense and observe its output.



```python
add_expense()
print("Called 'add_expense' function again to test the warning fix.")
```

## View All Expenses

### Subtask:
Develop a function to read and display all entries from the `expenses.csv` file in a clear, tabular format.


**Reasoning**:
I need to define the `view_all_expenses` function as described in the instructions, which will read `expenses.csv`, handle file not found errors, check for empty data, and display the expenses in a tabular format.



```python
def view_all_expenses():
    try:
        # Define dtypes to ensure consistency and avoid FutureWarning
        dtype_mapping = {'date': str, 'amount': float, 'category': str}
        df_expenses = pd.read_csv('expenses.csv', dtype=dtype_mapping)

        if df_expenses.empty:
            print("No expenses recorded yet.")
        else:
            print("\nAll Expenses:\n")
            print(df_expenses.to_string(index=False))
    except FileNotFoundError:
        print("Error: 'expenses.csv' not found. Please ensure it's initialized.")

print("Defined 'view_all_expenses' function.")
```

**Reasoning**:
The `view_all_expenses` function has been defined. Now, I need to call this function to demonstrate its functionality by displaying all existing expenses from the `expenses.csv` file.



```python
view_all_expenses()
print("Called 'view_all_expenses' function.")
```

## View Expenses by Category

### Subtask:
Create a function that prompts the user for a specific category, then reads `expenses.csv`, filters the entries by that category, and displays the filtered results.


**Reasoning**:
I need to define the `view_expenses_by_category` function that prompts the user for a category, reads `expenses.csv`, filters the expenses by the specified category in a case-insensitive manner, and then displays the results in a tabular format, handling various scenarios like file not found or no matching expenses.



```python
def view_expenses_by_category():
    category_input = input("Enter category to view: ")

    try:
        dtype_mapping = {'date': str, 'amount': float, 'category': str}
        df_expenses = pd.read_csv('expenses.csv', dtype=dtype_mapping)

        if df_expenses.empty:
            print("No expenses recorded yet.")
            return

        # Convert both input and DataFrame category to lowercase for case-insensitive matching
        filtered_df = df_expenses[df_expenses['category'].str.lower() == category_input.lower()]

        if filtered_df.empty:
            print(f"No expenses found for category: {category_input}")
        else:
            print(f"\nExpenses for category: {category_input}\n")
            print(filtered_df.to_string(index=False))

    except FileNotFoundError:
        print("Error: 'expenses.csv' not found. Please ensure it's initialized.")

print("Defined 'view_expenses_by_category' function.")
```

**Reasoning**:
The `view_expenses_by_category` function has been defined. To fulfill the subtask requirements, I need to call this function to demonstrate its functionality by prompting the user for a category and displaying the filtered results from the `expenses.csv` file.



```python
view_expenses_by_category()
print("Called 'view_expenses_by_category' function to test its functionality.")
```

## Create Expense Tracker Menu

### Subtask:
Build an interactive menu system that allows the user to choose between adding an expense, viewing all expenses, viewing expenses by category, or exiting the program. This will serve as the main interface for the expense tracker.


**Reasoning**:
I need to define the `expense_tracker_menu` function as instructed, which will display menu options, handle user input to call the appropriate expense tracking functions, and manage program exit and invalid inputs.



```python
def expense_tracker_menu():
    while True:
        print("\nExpense Tracker Menu:")
        print("1. Add new expense")
        print("2. View all expenses")
        print("3. View expenses by category")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_expense()
        elif choice == '2':
            view_all_expenses()
        elif choice == '3':
            view_expenses_by_category()
        elif choice == '4':
            print("Exiting Expense Tracker. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

print("Defined 'expense_tracker_menu' function.")
```

**Reasoning**:
The `expense_tracker_menu` function has been defined. To complete the subtask, I need to call this function to start the interactive expense tracker program as per the instructions.



```python
expense_tracker_menu()
print("Called 'expense_tracker_menu' to start the interactive program.")
```
